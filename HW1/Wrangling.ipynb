```python
# TODO: Create dataframe of airline_iata, number of incidents, number of routes
#routes_df
airline_incidents = routes_df.groupby(['airline_iata']).size().reset_index()
airline_incidents.columns = ['airline_iata', 'route_counts']
airline_incidents = airline_incidents.sort_values(['route_counts'], ascending = False)

new_incidents = pd.merge(incidents_df,
                         airlines_df, 
                         left_on='Airline', 
                         right_on='airline_name')[['Date','airline_name','airline_iata']]

incidents_routes = pd.merge(airline_incidents,
                            new_incidents, 
                            left_on='airline_iata', 
                            right_on='airline_iata')[['Date','airline_name','airline_iata','route_counts']]

incidents_routes = incidents_routes.groupby(['airline_iata']).size().reset_index()
incidents_routes.columns = ['airline_iata', 'incident_counts']
incidents_routes = incidents_routes.sort_values(['incident_counts'], ascending = False)

incidents_routes = pd.merge(incidents_routes,
                            airline_incidents, 
                            left_on='airline_iata', 
                            right_on='airline_iata')[['airline_iata','route_counts','incident_counts']]



incidents_routes
#routes_df
```

```python
# TODO: get random sample of flights from SQL
sample_flights_s_df = pd.read_sql_query('SELECT * FROM flights ORDER BY RANDOM() LIMIT 1000', con=engine)
sample_flights_s_df
```

```python
# TODO: create nested map of Jaccard distance

import numpy as np
def jaccard_dist(a,b):
    inter = np.intersect1d(a,b)
    union = np.union1d(a,b)
    return float(len(inter))/float(len(union))

#routes_df.dtypes
#flights_s_df.dtypes
#flights_s_df.select_dtypes(include=['object'])
```

## Step 5.1

```python
# TODO: output nested map
new_routes = routes_df.select_dtypes(['object'])
#I dont know if i should drop the duplicates because each entry is a different route
#new_routes = new_routes.drop_duplicates('airline_iata')

new_flights = sample_flights_s_df.select_dtypes(['object'])

nested_map = {}
inner_map = {}

for route_key in new_routes:
    for flight_key in new_flights:
        distance = jaccard_dist(new_routes[route_key],new_flights[flight_key])
        if distance != 0.0:
            inner_map[flight_key] = distance
        nested_map[route_key] = inner_map
    inner_map = {}
nested_map
```
